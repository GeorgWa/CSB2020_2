Grundlegende Idee für Parameter Tests: 

Es wird ein ausgangspunkt für die Parameter gewählt (Beispielsweise 5)
Pro Parameter werden X verschiedene Datensätze generiert.
Anschließend werden 1/2 X kleiner 5 und 1/2 X größer 5 generiert.
Der Abstand zwischen den Schritten kann durch einen parameter festgelegt werden.

Es wird jeweils der Unterschied zum Ausgangswert gebildet.
Der Unterschied wird durch den Veränderungsfaktor geteilt (z.B. 6/5)
Es wird der Mittelwert zwischen den einzelnen Werten gebildet. 
Dieser Mittlere Diff stellt dann den Einfluss auf die Kurve dar.


```python
def testFunc(a, b):
    res = []
    for i in range(10):
        res.append(a*(np.power((i-4.5), 2))+b)
    return res
```

```python
graph1=testFunc(1, 0)
graph2=testFunc(1.5, 0)
graph3=testFunc(0.5, 0)
print(graph3[4])
plt.plot(time, graph1, label="a=1")
plt.plot(time, graph2, label="a=2")
plt.plot(time, graph3, label="a=0.5")
plt.legend()
plt.show()
```

- "The basic reproductive number of an infection, R0, is defined as the expected number of secondary infectious cases generated by an average infectious case in an entirely susceptiblepopulation. This quantity determines the potential for an infectious agent to start an outbreak, the extent of transmission in the absence of control measures, and the ability of controlmeasures to reduce spread."
